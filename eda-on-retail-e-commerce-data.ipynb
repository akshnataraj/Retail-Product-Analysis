```python
# specify encoding to deal with different formats
df = pd.read_csv('../input/ecommerce-data/data.csv', encoding = 'ISO-8859-1')
df.head()
```

Information about Data:

```python
#check info on data
df.info()
# check missing values for each column 
df.isnull().sum().sort_values(ascending=False)
```

There are some null values in CustomerID and Description. Remove all rows with null values.
> CustomerID   406829

```python
# check missing values for each column 
reduced_df = df.dropna()
reduced_df.info()
reduced_df.isnull().sum().sort_values(ascending=False)
```

```python
Convert Data types of columns:
```

```python

# change CustomerID data type from string to int
reduced_df['CustomerID'] = reduced_df['CustomerID'].astype('int64')
# change the invoice_date format - String to Timestamp format
reduced_df['InvoiceDate'] = pd.to_datetime(reduced_df.loc['InvoiceDate'], format='%m/%d/%Y %H:%M')
```
